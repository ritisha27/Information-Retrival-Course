##Jaccard Coefficient

```python
DATA_DIRECTORY = 'content/CSE508_Winter2023_Dataset'
```

```python
# function to read a file.
def file_read(file_path):
  with open(file_path, "r") as file:
    data = file.read()
    
  return data
```

```python
# utility method to process the input query
def preprocess_input(input_str):
  input_str = input_str.lower()
  input_str = re.sub(r'[^\w\s]', '', input_str).split()

  without_stopwords = []
  for word in input_str:
    if word not in stop_words:
      without_stopwords.append(word)

  return without_stopwords
```

```python
# checking the correctness of the function
file_read('content/CSE508_Winter2023_Dataset/cranfield0001')
```

```python
# making a dictionary that stores docID and its content.
complete_file_content = dict()
files = sorted(os.listdir(DATA_DIRECTORY))

# enumerating over all the files in the given directory and store 
# the key as docID (index) and the value as the content of the file
for index, file in enumerate(files):
  file_path =  DATA_DIRECTORY + '/' + file
  data = file_read(file_path)
  complete_file_content[index+1] = data

# splitting the content of the file, so to make tokens
for i in range(1, len(complete_file_content)+1):
  complete_file_content[i] = complete_file_content[i].split()
```

```python
# function to calculate Jaccard Coefficient
def calculateJC(query,file_content):
  query = set(query)
  file_content = set(file_content)
  intersection_len = len(file_content.intersection(query))
  union_len = len(file_content.union(query))
  return (intersection_len/union_len)
```

```python
# utility function to get top 10 documents.
def get_top_documents():
  jc_index_file = dict()
  for i in range(1400):
    file_content = complete_file_content[i+1]
    jc= calculateJC(query,file_content)
    if(jc):
      jc_index_file[i+1] = jc                         # jc = key && docid=value

    # sort on the basis of jc values in decreasing value.
    jc_index_file = dict(sorted(jc_index_file.items(), key = lambda x: x[1], reverse = True))

    # findind top 10 documents
    top_10_docs = []
    for i in range(len(jc_index_file)):
      top_10_docs.append(list(jc_index_file.keys())[i])
      if(i==10):
        break

  return top_10_docs
```

####Main Function to run

```python
# Main function to run
query = input("Please Enter your query: ")
query = preprocess_input(query)
top_ten_docs = get_top_documents()

print()
print("Top Documents w.r.t to the given query: ")
print(top_ten_docs)
print()
for doc in top_ten_docs:
    print("Document No:  "+str(doc))
```
